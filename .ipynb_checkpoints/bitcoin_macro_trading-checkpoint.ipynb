```python
from typing import Optional, List

import numpy as np
import pandas as pd
import seaborn as sn
import statsmodels.api as sm
import requests
from loguru import logger
import ccxt
import time
from sklearn.linear_model import Ridge
```

## Load the price and interest rate datasets

```python
_binance = ccxt.binanceusdm()
_TARGET_COIN_TO_LIST = [
    'bitcoin',
]
```

```python
def load_ir_data() -> Optional[pd.DataFrame]:
    ir_df = pd.read_csv('United States 30-Year Bond Yield Historical Data.csv', index_col=['Date'], dayfirst=True, parse_dates=['Date'])
    ir_df = ir_df[['Price']].rename(columns={'Price': 'ir'})
    return ir_df
```

```python
def fetch_price_usd_daily_all(coin_id: str = 'bitcoin') -> Optional[pd.DataFrame]:
    url = 'https://api.coingecko.com/api/v3/coins/' + coin_id + \
    '/market_chart?vs_currency=usd&days=365&interval=daily'
    response = requests.get(url)
    data_dict = response.json()
    price_list = data_dict.get('prices')
    if price_list is None:
        logger.error(f'cannot fetch data from api, coin_id={coin_id}')
        return
    data_df = pd.DataFrame(price_list)
    data_df.columns = ['ts', coin_id + '_price_usd']
    data_df['ts'] = data_df['ts'].apply(lambda x: pd.to_datetime(_binance.iso8601(x)))
    data_df = data_df.set_index('ts')
    return data_df
```

```python
def download_coins_price(coin_id_list: List[str] = _TARGET_COIN_TO_LIST,
                         save_path: str = 'coins_price_data.xlsx') -> Optional[pd.DataFrame]:
    first_coin_id = coin_id_list[0]
    df_init = fetch_price_usd_daily_all(first_coin_id)
    if df_init is None:
        logger.warning(f'no data for coin, id={first_coin_id}')
        return
    merged_df: pd.DataFrame = pd.DataFrame(index=df_init.index)
    for coin_id in coin_id_list:
        time.sleep(20)
        logger.info(f'fetch price data, id={coin_id}')
        df = fetch_price_usd_daily_all(coin_id)
        if df is None:
            logger.warning(f'no data for coin, id={coin_id}')
        merged_df = pd.merge(merged_df, df, how='left', left_index=True, right_index=True)
    
    merged_df = merged_df.dropna()
    merged_df.index = merged_df.index.date
    merged_df.to_excel(save_path)
    return merged_df
```

```python
price_df = download_coins_price()
ir_df = load_ir_data()
```

```python
dfm = pd.merge(price_df, ir_df, how='left', left_index=True, right_index=True)
```

```python
dfm.dropna(inplace=True)
#dfm['ir'] = dfm['ir'].ffill()
```

```python
dfm['bitcoin_price_usd_pct_chg'] = dfm['bitcoin_price_usd'].pct_change()
dfm['ir_pct_chg'] = dfm['ir'].pct_change()
dfm['ir_pct_chg_t-1'] = dfm['ir_pct_chg'].shift()
dfm.dropna(inplace=True)
```

## Ridge Regression

Although the features are somewhat disappointing, we move forward to see how good we can get from them.

We use the ridge regression to predict the bitcoin price change of the next day. If positive, long; if negative; short.

But we want to make the backtesting of this simple trading strategy bit more realistic by splitting the whole dataset into three parts:
1. First 70%: training data; for training and hyperparameter tuning
2. Middle 10%: leave this out and not do anything; this is for breaking the temporal dependence between the training data and the test data
3. Last 20%: test data; for evaluating the trading strategy

```python
n_alpha = 50
alphas = np.logspace(-2, 4, n_alpha)
ridge_res = {}
```

```python
N = len(dfm.index)

train_ratio = 0.6
val_ratio = 0.2
train_end = int(N * train_ratio)
val_end = train_end + int(N * val_ratio)

split_idx = [
    dfm.index[:train_end],
    dfm.index[train_end:val_end],
    dfm.index[val_end:]
]

train_df = dfm.loc[split_idx[0]]
test_df = dfm.loc[split_idx[2]]
```

```python
hitratio_arr = np.array([])
for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(train_df[['ir_pct_chg_t-1', 'Active market wallets_pct_chg_t-1', 'Sales USD_pct_chg_t-1']],
              train_df[['bitcoin_price_usd_pct_chg']])
    hitratio = np.mean(ridge.predict(train_df[['ir_pct_chg_t-1', 'Active market wallets_pct_chg_t-1', 'Sales USD_pct_chg_t-1']])*train_df[['bitcoin_price_usd_pct_chg']] > 0)
    hitratio_arr = np.append(hitratio_arr, hitratio)
```

```python
best_alpha = alphas[np.argmax(hitratio_arr)]
print(f'best alpha: {best_alpha}')
```

```python
ridge = Ridge(alpha=best_alpha)
ridge.fit(train_df[['ir_pct_chg_t-1', 'Active market wallets_pct_chg_t-1', 'Sales USD_pct_chg_t-1']],
          train_df[['bitcoin_price_usd_pct_chg']])
hitratio = np.mean(ridge.predict(train_df[['ir_pct_chg_t-1', 'Active market wallets_pct_chg_t-1', 'Sales USD_pct_chg_t-1']])*train_df[['bitcoin_price_usd_pct_chg']] > 0)
print(f'hit ratio applied to the test data: {hitratio}')
```

... well, a bit better than flipping a coin, and it is very likely this would vanish due to transaction fee.

## Strategy Evaluation

```python
test_df['predicted_bitcoin_price_usd_pct_chg'] = ridge.predict(test_df[['ir_pct_chg_t-1', 'Active market wallets_pct_chg_t-1', 'Sales USD_pct_chg_t-1']])
test_df['position'] = np.where(test_df['predicted_bitcoin_price_usd_pct_chg'] > 0, 1, -1)
```

```python
test_df['pnl'] = test_df['position'].shift() * test_df['bitcoin_price_usd_pct_chg']
test_df['cumul_pnl'] = test_df['pnl'].cumsum()
test_df['benchmark'] = test_df['bitcoin_price_usd_pct_chg'].cumsum()
```

```python
test_df[['cumul_pnl', 'benchmark']].plot();
```

```python
def compute_SR(r_data: pd.Series) -> float:
    return r_data.mean() / r_data.std() * np.sqrt(365)
```

```python
print('strat SR ' + str(compute_SR(test_df['pnl'])))
print('benchmark SR ' + str(compute_SR(test_df['bitcoin_price_usd_pct_chg'])))
```

... Overall, this strategy is far from satisfactory. If you change the ratios of the training and test data, you would get a very different performance, showing this strategy is not reliable. Also the Sharpe ratio is < 1, meaning that it does not really take risks smartly.

Having said that, we have demonstrated in this notebook how to backtest a trading idea - not bad at all!
